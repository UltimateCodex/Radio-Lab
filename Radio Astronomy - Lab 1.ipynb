```python
import ugradio
import numpy as np
import argparse
import time
import matplotlib.pylab as plt
% matplotlib inline

# 3.2
# N = 5

samples_used = 50
v_samp = 12.5e6 # Hz

times = np.linspace((-samples_used/2)/v_samp, ((samples_used/2)-1)/v_samp, samples_used)/(1e-6)

frequencies2 = ["1.25", "2.5", "3.75", "5", "6.25", "7.5", "10", "11.25", "13.75"]

for f in frequencies2:
    data = np.load(f + "MHz" + "_1_.npy")
    voltage = data[:samples_used]
    plt.plot(times, voltage, 'k-')
    plt.xlabel("Time ($\mu$ seconds)")
    plt.ylabel("Voltage (V)")
    plt.title(f + "MHz signal sampled at 12.5 MHz")
    plt.xlim((times[0], times[-1]))
    plt.show()
```
